```python
import numpy as np
import matplotlib.pyplot as plt

# Define timeline with 100 points between -π and π
timeline = np.linspace(-np.pi, np.pi, 100)

# Compute sine values
sin_values = np.sin(timeline)

# Plot
plt.plot(timeline, sin_values, label=r'$y = \sin(x)$', color='b')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Sine Wave")
plt.axhline(0, color='black', linewidth=0.5)  # X-axis
plt.axvline(0, color='black', linewidth=0.5)  # Y-axis
plt.grid(True, linestyle='--', alpha=0.6)  # Grid
plt.legend()
plt.show()
```
